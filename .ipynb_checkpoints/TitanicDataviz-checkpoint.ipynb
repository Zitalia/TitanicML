```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_clean=pd.read_csv('data/train_clean.csv')
test_clean=pd.read_csv('data/test_clean.csv')
```

```python
explode = (0, 0.1)
fig1, ax1 = plt.subplots()
ax1.pie(train_clean['Survived'].value_counts(), explode=explode, labels=['Naufrag√©s', 'Survivants'], autopct='%1.1f%%',
        shadow=True, startangle=90)
```

```python
plt.scatter(train_clean['Age'], train_clean['Survived'])
plt.ylabel("Survivant")
plt.xlabel("Age")
```

```python
plt.scatter(train_clean['Fare'], train_clean['Survived'])
plt.ylabel("Survivant")
plt.xlabel("Tarifs")
```

```python
#Le sexe survivant
plt.hist(train_clean['Sex'])
plt.hist(train_clean['Sex'].where(train_clean['Survived'] == 1).dropna(), color="r")
```

```python
fem = train_clean.where( train_clean['Sex']=='female').dropna()
hom = train_clean.where( train_clean['Sex']=='male').dropna()
plt.hist([fem['Survived'],hom['Survived']])
plt.legend(['femme','homme'])
```
